# Support Vector Machines SVM

## simulate data

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt


blobs = make_blobs([300,300], 2, random_state=42, centers=[(-2,-2), (2,2)])

blobs
```

```python
blobs[0]
```

```python
blobs[1]
```

```python
blobs_data = blobs[0]
blobs_label = blobs[1]
```

```python
plt.scatter(blobs_data[:,0], blobs_data[:,1])
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

blobs_with_label = np.c_[blobs_data, blobs_label]
```

```python
df = pd.DataFrame(blobs_with_label, columns=['X1', 'X2', 'label'])
sns.scatterplot(data=df, x='X1', y='X2', hue='label')
```

```python
X,y= blobs_data, blobs_label
X[:5], y[:5]
```

## SVM

### Linear kernel

```python
from utils import plot_svm_margins
#support vector classifier
from sklearn.svm import SVC

fig, axes = plt.subplots(1,3, figsize=(16,4))

# C is the regularisation parameter (budget)
# Makes marger sofetr
for i, C in enumerate([10,0.1,0.01], 1):
    plt.subplot(1,3,i)
    plot_svm_margins(SVC(kernel='linear', C=C), X=X, y=y)
    plt.title(f"C = {C}, linear kernel")
    
# when using linear kernel, rather use LinearSVC
```

### Radial basis kernel

```python
from utils import plot_svm_margins
#support vector classifier
from sklearn.svm import SVC

fig, axes = plt.subplots(1,3, figsize=(16,4))

# C is the regularisation parameter (budget)
# Makes marger sofetr
for i, C in enumerate([10,1, 0.1], 1):
    plt.subplot(1,3,i)
    plot_svm_margins(SVC(kernel='rbf', C=C), X=X, y=y)
    plt.title(f"C = {C}, radial basis function kernel")
```

### Polynomial kernel

```python
from utils import plot_svm_margins
#support vector classifier
from sklearn.svm import SVC

fig, axes = plt.subplots(1,3, figsize=(16,4))

# C is the regularisation parameter (budget)
# Makes marger sofetr
for i, C in enumerate([10,1, 0.1], 1):
    plt.subplot(1,3,i)
    plot_svm_margins(SVC(kernel='poly', C=C, degree=5), X=X, y=y)
    plt.title(f"C = {C}, polynomial kernel")
```
